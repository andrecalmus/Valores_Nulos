```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import glob
```

```python
def cargar_archivo(dataframe):
    if dataframe.endswith('.csv'):
        data = pd.read_csv(dataframe)
    elif dataframe.endswith('.xlsx'):
        data = pd.read_xlsx(dataframe)
    elif dataframe.endswith('.html'):
        data = pd.read_html(dataframe)
    elif dataframe.endswith('.json'):
        with open(dataframe) as f:
            data = json.load(f)
    else:
        print("Tipo de archivo inválido, favor de corrobar que haya cargado un archivo .csv, .xlsx, .html o .json")  
    return(data)
```

Prueba para corroborar que sirva nuestra función, se cargará un archivo .csv

```python
cargar_archivo("GINEBRA.csv")
```
