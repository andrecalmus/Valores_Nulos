Generar una función para cada una de las siguientes acciones:
a) Reemplazar valores nulos con el método de “mediana” en las columnas tipo “numeric (int y float)”
b) Reemplazar valores nulos con el método de “ffill” en las columnas tipo “object, datetime y category”
c) Reemplazar valores nulos con el método de “bfill” en las columnas tipo “object, datetime y category”
d) Reemplazar valores nulos con el método de “string concreto” en las columnas tipo “object, datetime y category”
e) Reemplazar valores nulos con el método de “constante” en las columnas tipo “numeric (int y float)”

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Convertir a dataframe
df = pd.read_csv("GINEBRA.csv")
df
```

```python
#Eliminar columnas innecesarias
#axis = 1 columnas
#axis = 0 filas
df1 = df.drop(["Unnamed: 0"], axis=1)
df1
```

```python
#Identificar valores nulos por columna
valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
df2 = df1.copy()
```

```python
def sustitucion_mean(dataframe):
    cuantitativas_con_nulos = dataframe.select_dtypes(include=['float64', 'int64','float','int'])
    cualitativas = dataframe.select_dtypes(include=['object', 'datetime','category'])
    cuantitativas = cuantitativas_con_nulos.fillna(round(cuantitativas_con_nulos.mean(), 1))
    Datos_sin_nulos = pd.concat([cuantitativas, cualitativas], axis=1)
    return(Datos_sin_nulos)

def sustitucion_ffill(dataframe):
    cuantitativas = dataframe.select_dtypes(include=['float64', 'int64','float','int'])
    cualitativas_con_nulos = dataframe.select_dtypes(include=['object', 'datetime','category'])
    cualitativas = cualitativas_con_nulos.fillna(method='ffill')
    Datos_sin_nulos = pd.concat([cuantitativas, cualitativas], axis=1)
    return(Datos_sin_nulos)

def sustitucion_bfill(dataframe):
    cuantitativas = dataframe.select_dtypes(include=['float64', 'int64','float','int'])
    cualitativas_con_nulos = dataframe.select_dtypes(include=['object', 'datetime','category'])
    cualitativas = cualitativas_con_nulos.fillna(method='bfill')
    Datos_sin_nulos = pd.concat([cuantitativas, cualitativas], axis=1)
    return(Datos_sin_nulos)


def sustitucion_string_concreto(dataframe,cadena):
    cuantitativas = dataframe.select_dtypes(include=['float64', 'int64','float','int'])
    cualitativas_con_nulos = dataframe.select_dtypes(include=['object', 'datetime','category'])
    cualitativas = cualitativas_con_nulos.fillna(cadena)
    Datos_sin_nulos = pd.concat([cuantitativas, cualitativas], axis=1)
    return(Datos_sin_nulos)

def sustitucion_constante(dataframe,constante):
    cuantitativas_con_nulos = dataframe.select_dtypes(include=['float64', 'int64','float','int'])
    cualitativas = dataframe.select_dtypes(include=['object', 'datetime','category'])
    cuantitativas = cuantitativas_con_nulos.fillna(constante)
    Datos_sin_nulos = pd.concat([cuantitativas, cualitativas], axis=1)
    return(Datos_sin_nulos)
```

```python
sustitucion_mean = sustitucion_mean(df2)
sustitucion_mean.to_csv('sustitucion_mean.csv')
sustitucion_mean.isnull().sum()
```

```python
sustitucion_ffill = sustitucion_ffill(df2)
sustitucion_ffill.to_csv('sustitucion_ffill.csv')
sustitucion_ffill.isnull().sum()
```

```python
sustitucion_bfill = sustitucion_bfill(df2)
sustitucion_bfill.to_csv('sustitucion_bfill.csv')
sustitucion_bfill.isnull().sum()
```

```python
sustitucion_string_concreto = sustitucion_string_concreto(df2,'a few days or more')
sustitucion_string_concreto.to_csv('sustitucion_string_concreto.csv')
sustitucion_string_concreto.isnull().sum()
```

```python
sustitucion_constante = sustitucion_constante(df2, 5)
sustitucion_constante.to_csv('sustitucion_constante.csv')
sustitucion_constante.isnull().sum()
```
